```python
from mp_api.client import MPRester
from pymatgen.core import Composition
import pandas as pd

candidate_formulas = df.formula.tolist()

mp_records = []

with MPRester(MP_API_KEY) as mpr:
    for formula in candidate_formulas:
        target_comp = Composition(formula).reduced_composition

        docs = mpr.materials.summary.search(
            formula=formula
        )

        exact_matches = []
        for d in docs:
            if d.composition_reduced == target_comp:
                exact_matches.append(d)

        if not exact_matches:
            formula_data = df[df["formula"] == formula].iloc[0]
            mp_records.append({
                "formula": formula,
                "A": formula_data.A,
                "B": formula_data.B,
                "B_prime" : formula_data.B_prime,
                "fitness" : formula_data.fitness,
                "predicted_band_gap" : formula_data.predicted_band_gap,
                "predicted_ehull" : formula_data.predicted_ehull,         
                "predicted_formation_energy":  formula_data.predicted_formation_energy,
                "charge_neutral" : formula_data.charge_neutral,
            })

novel_candidates= pd.DataFrame(mp_records)
novel_candidates.to_csv("ga_candidates_mp_existence_check.csv", index=False)

novel_candidates
```

```python
import numpy as np
from pymatgen.core.periodic_table import Species, Element

R_O = 1.40  # Shannon radius for O2- in octahedral coordination, approximate

OXIDATION_STATES_CHOSEN = {
    "Li": 1, "Na": 1, "K": 1, "Rb": 1, "Cs": 1,
    "Mg": 2, "Ca": 2, "Sr": 2, "Ba": 2,
    "La": 3, "Y": 3,

    "Al": 3, "Ga": 3, "In": 3,
    "Ti": 4, "V": 5, "Cr": 3, "Mn": 4,
    "Fe": 3, "Co": 3, "Ni": 3, "Cu": 2,
    "Zn": 2, "Zr": 4, "Hf": 4,
    "Nb": 5, "Ta": 5,
    "Mo": 6, "W": 6,
    "Sn": 4, "Sb": 5, "Bi": 3
}


def get_ionic_radius(element, oxidation_state, coordination):
    # First try Shannon radius with coordination number
    try:
        sp = Species(element, oxidation_state)
        r = sp.get_shannon_radius(cn=coordination)
        if r is not None:
            return r
    except Exception:
        pass

    # Fallback: use pymatgen ionic radii dictionary
    try:
        radii_dict = Element(element).ionic_radii
        r = radii_dict.get(oxidation_state)
        if r is not None:
            return float(r)
    except Exception:
        pass

    return np.nan


def tolerance_octahedral_factors(A, B, Bp):
    ox_A = OXIDATION_STATES_CHOSEN.get(A)
    ox_B = OXIDATION_STATES_CHOSEN.get(B)
    ox_Bp = OXIDATION_STATES_CHOSEN.get(Bp)

    if ox_A is None or ox_B is None or ox_Bp is None:
        return np.nan, np.nan

    # A-site: 12-fold coordination, B-sites: 6-fold coordination
    rA = get_ionic_radius(A, ox_A, 12)
    rB = get_ionic_radius(B, ox_B, 6)
    rBp = get_ionic_radius(Bp, ox_Bp, 6)

    if np.isnan(rA) or np.isnan(rB) or np.isnan(rBp):
        return np.nan, np.nan

    rB_avg = (rB + rBp) / 2

    t = (rA + R_O) / (np.sqrt(2) * (rB_avg + R_O))
    mu = rB_avg / R_O

    return t, mu
```

```python
df_screened = novel_candidates.copy()

t_values = []
mu_values = []

for _, row in df_screened.iterrows():
    t, mu = tolerance_octahedral_factors(
        row["A"],
        row["B"],
        row["B_prime"]
    )
    t_values.append(t)
    mu_values.append(mu)

df_screened["tolerance_factor_t"] = t_values
df_screened["octahedral_factor_mu"] = mu_values

df_screened["passes_geometry_screen"] = (
    (df_screened["tolerance_factor_t"] >= 0.80) &
    (df_screened["tolerance_factor_t"] <= 1.05) &
    (df_screened["octahedral_factor_mu"] >= 0.414) &
    (df_screened["octahedral_factor_mu"] <= 0.732)
)

df_screened.to_csv("ga_candidates_with_geometry_screen.csv", index=False)

df_screened[
    [
        "formula",
        "fitness",
        "predicted_band_gap",
        "predicted_ehull",
        "tolerance_factor_t",
        "octahedral_factor_mu",
        "passes_geometry_screen"
    ]
].head(36)
```

```python
# Chemical confirmation
PREFERRED_OXIDATION_STATES = {
    "Li": [1], "Na": [1], "K": [1], "Rb": [1], "Cs": [1],
    "Mg": [2], "Ca": [2], "Sr": [2], "Ba": [2],
    "La": [3], "Y": [3],

    "Al": [3], "Ga": [3], "In": [3],
    "Ti": [4],
    "V": [3, 4, 5],
    "Cr": [3, 4],
    "Mn": [2, 3, 4],
    "Fe": [2, 3],
    "Co": [2, 3],
    "Ni": [2, 3],
    "Cu": [1, 2],
    "Zn": [2],
    "Zr": [4],
    "Hf": [4],
    "Nb": [5],
    "Ta": [5],
    "Mo": [4, 5, 6],
    "W": [5, 6],
    "Sn": [4],
    "Sb": [5],
    "Bi": [3, 5]
}

SUSPICIOUS_OXIDATION_STATES = {
    ("Ni", 4),
    ("Cu", 3),
    ("Co", 4),
    ("Fe", 4),
    ("Cr", 6),
    ("Mn", 7),
    ("Sn", 2),
    ("Sb", 3),
    ("Bi", 5)
}
```

```python
def get_plausible_oxidation_combinations(A, B, Bp):
    combos = []

    for ox_A in PREFERRED_OXIDATION_STATES.get(A, []):
        for ox_B in PREFERRED_OXIDATION_STATES.get(B, []):
            for ox_Bp in PREFERRED_OXIDATION_STATES.get(Bp, []):

                if 2 * ox_A + ox_B + ox_Bp != 12:
                    continue

                suspicious_count = 0

                if (A, ox_A) in SUSPICIOUS_OXIDATION_STATES:
                    suspicious_count += 1
                if (B, ox_B) in SUSPICIOUS_OXIDATION_STATES:
                    suspicious_count += 1
                if (Bp, ox_Bp) in SUSPICIOUS_OXIDATION_STATES:
                    suspicious_count += 1

                combos.append({
                    "ox_A": ox_A,
                    "ox_B": ox_B,
                    "ox_B_prime": ox_Bp,
                    "suspicious_count": suspicious_count
                })

    return combos
```

```python
plausibility_results = []

for _, row in df_screened.iterrows():
    combos = get_plausible_oxidation_combinations(
        row["A"],
        row["B"],
        row["B_prime"]
    )

    if len(combos) == 0:
        plausibility_results.append({
            "has_plausible_oxidation": False,
            "best_ox_A": None,
            "best_ox_B": None,
            "best_ox_B_prime": None,
            "suspicious_count": None
        })
    else:
        best_combo = sorted(combos, key=lambda x: x["suspicious_count"])[0]

        plausibility_results.append({
            "has_plausible_oxidation": best_combo["suspicious_count"] == 0,
            "best_ox_A": best_combo["ox_A"],
            "best_ox_B": best_combo["ox_B"],
            "best_ox_B_prime": best_combo["ox_B_prime"],
            "suspicious_count": best_combo["suspicious_count"]
        })

plausibility_df = pd.DataFrame(plausibility_results)

df_screened = pd.concat(
    [df_screened.reset_index(drop=True), plausibility_df],
    axis=1
)
```

```python
df_dft_ready = df_screened[
    (df_screened["passes_geometry_screen"] == True) &
    (df_screened["has_plausible_oxidation"] == True)
].copy()

df_dft_ready = df_dft_ready.sort_values(
    by=["fitness", "predicted_ehull"],
    ascending=[True, True]
).reset_index(drop=True)

df_dft_ready[
    [
        "formula",
        "fitness",
        "predicted_band_gap",
        "predicted_ehull",
        "tolerance_factor_t",
        "octahedral_factor_mu",
        "best_ox_A",
        "best_ox_B",
        "best_ox_B_prime"
    ]
].head(17)
```
