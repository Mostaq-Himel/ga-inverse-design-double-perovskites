```python
# A-site pool (alkaline-earth + rare-earth)
A_SITE = [
    "Ba", "Sr", "Ca", "Mg", "Li", "Na", "K", "Rb", "Cs",
    "La", "Ce", "Pr", "Nd", "Sm", "Gd", "Y", "Bi"
]

# B-site pool (ransition metals for perovskites)
B_SITE = [
    "Fe", "Mn", "Co", "Ni", "Ti", "Cr", "V", "Cu", "Zn", "Mo", "W"
]

# B' site pool (ifferent transition metals)
B_PRIME_SITE = [
    "Nb", "Ta", "Mo", "W", "V", "Ti", "Cr", "Re"
]
```

```python
# Structural penalty function
def tolerance_factor(A, B, Bp):
    try:
        rA = IONIC_RADII[A]
        rB = (IONIC_RADII[B] + IONIC_RADII[Bp]) / 2
        rO = 1.40
        t = (rA + rO) / (1.414 * (rB + rO))
        return t
    except:
        return None

def octahedral_factor(B, Bp):
    try:
        rB = (IONIC_RADII[B] + IONIC_RADII[Bp]) / 2
        rO = 1.40
        mu = rB / rO
        return mu
    except:
        return None

def structural_penalty(A, B, Bp):
    t = tolerance_factor(A, B, Bp)
    mu = octahedral_factor(B, Bp)
    if t is None or mu is None:
        return 10.0  # invalid
    penalty = 0.0
    if not 0.8 <= t <= 1.0:
        penalty += abs(t-0.9)*10
    if not 0.414 <= mu <= 0.895:
        penalty += abs(mu-0.65)*10
    return penalty
```

```python
ep = ElementProperty.from_preset("magpie")
```

```python
# ML featurization function
from pymatgen.core import Composition
import pandas as pd

def featurize_formula(formula, feature_columns):
    """
    Convert formula into numeric features for RF model prediction.
    Ensures the same columns as the training set.
    """
    comp = Composition(formula)
    df = pd.DataFrame({"composition": [comp]})
    
    # Generate Magpie features
    X = ep.featurize_dataframe(df, col_id="composition", ignore_errors=True)
    
    # Keep only numeric columns
    X = X.select_dtypes(include="number")
    
    # Reindex to match training columns; missing columns get filled with 0
    X = X.reindex(columns=feature_columns, fill_value=0)
    
    return X
```

```python
# DEAP - NSGA-II setup
# Multi-objective fitness: maximize Eg, minimize penalty
creator.create("FitnessMulti", base.Fitness, weights=(1.0, -1.0))
creator.create("Individual", list, fitness=creator.FitnessMulti)

toolbox = base.Toolbox()
toolbox.register("attr_A", random.randrange, len(A_SITE))
toolbox.register("attr_B", random.randrange, len(B_SITE))
toolbox.register("attr_Bp", random.randrange, len(B_PRIME_SITE))
toolbox.register("individual", tools.initCycle, creator.Individual,
                 (toolbox.attr_A, toolbox.attr_B, toolbox.attr_Bp), n=1)
toolbox.register("population", tools.initRepeat, list, toolbox.individual)
toolbox.register("mate", tools.cxUniform, indpb=0.5)
toolbox.register("mutate", tools.mutUniformInt, 
                 low=[0,0,0], up=[len(A_SITE)-1,len(B_SITE)-1,len(B_PRIME_SITE)-1], indpb=0.2)
toolbox.register("select", tools.selNSGA2)
```

```python
# Fitness evaluation using ML+structural penalty

def evaluate_nsga(ind):
    A = A_SITE[ind[0]]
    B = B_SITE[ind[1]]
    Bp = B_PRIME_SITE[ind[2]]

    if B == Bp:
        return (0.0, 10.0)

    formula = f"{A}2{B}{Bp}O6"

    try:
        X = featurize_formula(formula, FEATURE_COLUMNS)  # ✅ pass FEATURE_COLUMNS here
        eg = rf_model.predict(X)[0]
        if eg < 0 or eg > 6:
            return (0.0, 10.0)
    except Exception as e:
        print(f"Prediction failed for {formula}: {e}")
        return (0.0, 10.0)

    penalty = structural_penalty(A, B, Bp)
    return (eg, penalty)


toolbox.register("evaluate", evaluate_nsga)
```

```python
# NSGA-II execution 
POP_SIZE = 120
N_GEN = 10

# Initialize population
pop = toolbox.population(n=POP_SIZE)
print(len(pop))

# Evaluate initial population
for ind in pop:
    ind.fitness.values = toolbox.evaluate(ind)

# Evolutionary loop
for gen in range(N_GEN):
    offspring = algorithms.varAnd(pop, toolbox, cxpb=0.6, mutpb=0.3)
    print("running")

    for ind in offspring:
        ind.fitness.values = toolbox.evaluate(ind)
    pop = toolbox.select(pop + offspring, k=POP_SIZE)

    # Pareto front of current generation
    front = tools.sortNondominated(pop, k=len(pop), first_front_only=True)[0]
    best = front[0]
    print( best)

    if best.fitness.valid:
        A, B, Bp = A_SITE[best[0]], B_SITE[best[1]], B_PRIME_SITE[best[2]]
        print(
            f"Gen {gen:02d} | "
            f"Eg={best.fitness.values[0]:.2f} eV | "
            f"penalty={best.fitness.values[1]:.3f} | "
            f"{A}2{B}{Bp}O6"
        )
```
