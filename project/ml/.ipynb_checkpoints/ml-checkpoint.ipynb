```python
import pandas as pd
import numpy as np

from mp_api.client import MPRester
from pymatgen.core import Composition

from matminer.featurizers.composition import (
    ElementProperty,
    Stoichiometry,
    IonProperty
)

from sklearn.preprocessing import StandardScaler

```

```python
MP_API_KEY = "nvpzeRRBghdu6Cowj8XxwJHNmToCES32"
```

```python
# Query data from MP
with MPRester(MP_API_KEY) as mpr:
    results = mpr.materials.summary.search(
        # formula="*O6",
        elements=["O"],
        num_elements=4,
        energy_above_hull=(0, 0.05),
        fields=[
            "material_id",
            "formula_pretty",
            "structure",
            "band_gap",
            "energy_above_hull"
        ],
        chunk_size=1000
    )

df_raw = pd.DataFrame(results)
print("Total entries fetched:", len(df_raw))
df_raw.head()
```

```python
from pymatgen.core import Composition

def is_double_perovskite(formula):
    comp = Composition(formula)
    el_amt = comp.get_el_amt_dict()

    # Must have exactly 4 elements
    if len(el_amt) != 4:
        return False
        
    # Must contain exactly O6
    if el_amt.get("O", 0) != 6:
        return False

    # Non-oxygen atoms
    cations = {el: amt for el, amt in el_amt.items() if el != "O"}

    # Must match A2BB'
    values = sorted(cations.values())
    return values == [1, 1, 2]

df_dp = df_raw[df_raw["formula_pretty"].apply(is_double_perovskite)]
df_dp = df_dp.reset_index(drop=True)

print("Double perovskites found:", len(df_dp))
df_dp.head()
```
